```python
import torch
import torch.nn as nn
import numpy as np
from sklearn.metrics import mean_squared_error

# Convert to PyTorch tensors
X_train_tensor = torch.FloatTensor(X_lstm_train)
y_train_tensor = torch.FloatTensor(y_lstm_train)
X_test_tensor = torch.FloatTensor(X_lstm_test)
y_test_tensor = torch.FloatTensor(y_lstm_test)

# Define LSTM model
class LSTMModel(nn.Module):
    def __init__(self, input_size=1, hidden_size=64, num_layers=1):
        super(LSTMModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)
    
    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)
        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out

# Create and train model
model = LSTMModel()
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# Training loop
epochs = 100
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    output = model(X_train_tensor)
    loss = criterion(output.squeeze(), y_train_tensor.squeeze())
    loss.backward()
    optimizer.step()
    
    if epoch % 20 == 0:
        print(f'Epoch {epoch}, Loss: {loss.item():.4f}')
```

```python
import torch
import torch.nn as nn
import numpy as np
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Convert to PyTorch tensors
X_train_tensor = torch.FloatTensor(X_lstm_train)
y_train_tensor = torch.FloatTensor(y_lstm_train)
X_test_tensor = torch.FloatTensor(X_lstm_test)
y_test_tensor = torch.FloatTensor(y_lstm_test)

# Improved LSTM model
class ImprovedLSTMModel(nn.Module):
    def __init__(self, input_size=1, hidden_size=128, num_layers=2, dropout=0.2):
        super(ImprovedLSTMModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        
        # LSTM layers with dropout
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, 
                           batch_first=True, dropout=dropout if num_layers > 1 else 0)
        
        # Fully connected layers with dropout
        self.dropout = nn.Dropout(dropout)
        self.fc1 = nn.Linear(hidden_size, 64)
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 1)
        self.relu = nn.ReLU()
    
    def forward(self, x):
        # Initialize hidden states
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)
        
        # LSTM forward pass
        out, _ = self.lstm(x, (h0, c0))
        
        # Take the last output
        out = out[:, -1, :]
        
        # Fully connected layers
        out = self.dropout(out)
        out = self.relu(self.fc1(out))
        out = self.dropout(out)
        out = self.relu(self.fc2(out))
        out = self.fc3(out)
        
        return out

# Create model with better hyperparameters
model = ImprovedLSTMModel(input_size=1, hidden_size=128, num_layers=2, dropout=0.2)

# Better optimizer and loss function
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=10, factor=0.5)

# Training with early stopping
epochs = 500
best_loss = float('inf')
patience = 20
patience_counter = 0
train_losses = []
val_losses = []

print("Training improved LSTM model...")
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    
    # Forward pass
    output = model(X_train_tensor)
    train_loss = criterion(output.squeeze(), y_train_tensor.squeeze())
    
    # Backward pass
    train_loss.backward()
    
    # Gradient clipping to prevent exploding gradients
    torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
    
    optimizer.step()
    
    # Validation
    model.eval()
    with torch.no_grad():
        val_output = model(X_test_tensor)
        val_loss = criterion(val_output.squeeze(), y_test_tensor.squeeze())
    
    train_losses.append(train_loss.item())
    val_losses.append(val_loss.item())
    
    # Learning rate scheduling
    scheduler.step(val_loss)
    
    # Early stopping
    if val_loss < best_loss:
        best_loss = val_loss
        patience_counter = 0
        # Save best model
        best_model_state = model.state_dict()
    else:
        patience_counter += 1
    
    if patience_counter >= patience:
        print(f"Early stopping at epoch {epoch}")
        break
    
    if epoch % 50 == 0:
        print(f'Epoch {epoch}, Train Loss: {train_loss.item():.6f}, Val Loss: {val_loss.item():.6f}')

# Load best model
model.load_state_dict(best_model_state)

# Final evaluation
model.eval()
with torch.no_grad():
    train_pred = model(X_train_tensor).squeeze().numpy()
    test_pred = model(X_test_tensor).squeeze().numpy()

# Calculate RMSE
train_rmse = np.sqrt(mean_squared_error(y_lstm_train, train_pred))
test_rmse = np.sqrt(mean_squared_error(y_lstm_test, test_pred))

print(f"\nFinal Results:")
print(f"Train RMSE: {train_rmse:.6f}")
print(f"Test RMSE: {test_rmse:.6f}")

# Plot training curves
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.yscale('log')

plt.subplot(1, 2, 2)
plt.plot(y_lstm_test, label='Actual', alpha=0.7)
plt.plot(test_pred, label='Predicted', alpha=0.7)
plt.title('LSTM Predictions vs Actual')
plt.xlabel('Time')
plt.ylabel('Residuals')
plt.legend()

plt.tight_layout()
plt.show()

# Transform predictions back to original scale
train_pred_original = scaler.inverse_transform(train_pred.reshape(-1, 1)).flatten()
test_pred_original = scaler.inverse_transform(test_pred.reshape(-1, 1)).flatten()
y_train_original = scaler.inverse_transform(y_lstm_train.reshape(-1, 1)).flatten()
y_test_original = scaler.inverse_transform(y_lstm_test.reshape(-1, 1)).flatten()

# Calculate RMSE in original scale
train_rmse_original = np.sqrt(mean_squared_error(y_train_original, train_pred_original))
test_rmse_original = np.sqrt(mean_squared_error(y_test_original, test_pred_original))

print(f"\nOriginal Scale Results:")
print(f"Train RMSE: {train_rmse_original:.6f}")
print(f"Test RMSE: {test_rmse_original:.6f}")
```

```python
# CORRECT APPROACH: Avoid data leakage and check if residuals have patterns

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import torch
import torch.nn as nn

print("=== ANALYZING RESIDUALS FOR TEMPORAL PATTERNS ===")

# 1. CORRECT residual calculation (avoid data leakage)
# Only use training residuals for LSTM training
residual_train = y_train - xgb_pred_train
residual_test_true = y_test - xgb_pred_test  # Only for evaluation, not training

print(f"Residual statistics:")
print(f"Train residuals - Mean: {np.mean(residual_train):.4f}, Std: {np.std(residual_train):.4f}")
print(f"Test residuals - Mean: {np.mean(residual_test_true):.4f}, Std: {np.std(residual_test_true):.4f}")

# 2. Check if residuals have temporal patterns worth modeling
def analyze_residual_patterns(residuals, name):
    plt.figure(figsize=(15, 5))
    
    # Plot residuals over time
    plt.subplot(1, 3, 1)
    plt.plot(residuals[:200])
    plt.title(f'{name} - Time Series')
    plt.xlabel('Time')
    plt.ylabel('Residual')
    
    # Plot histogram
    plt.subplot(1, 3, 2)
    plt.hist(residuals, bins=50, alpha=0.7)
    plt.title(f'{name} - Distribution')
    plt.xlabel('Residual Value')
    plt.ylabel('Frequency')
    
    # Plot autocorrelation
    plt.subplot(1, 3, 3)
    from statsmodels.tsa.stattools import acf
    autocorr = acf(residuals, nlags=20, fft=True)
    plt.stem(range(len(autocorr)), autocorr)
    plt.axhline(y=0.05, color='red', linestyle='--', alpha=0.5, label='5% threshold')
    plt.axhline(y=-0.05, color='red', linestyle='--', alpha=0.5)
    plt.title(f'{name} - Autocorrelation')
    plt.xlabel('Lag')
    plt.ylabel('Autocorrelation')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    # Check if there are significant patterns
    significant_autocorr = np.sum(np.abs(autocorr[1:]) > 0.05)
    print(f"{name} - Significant autocorrelations: {significant_autocorr}/20")
    
    return significant_autocorr > 2  # If more than 2 significant lags, worth modeling

# Analyze patterns
has_patterns = analyze_residual_patterns(residual_train, "Training Residuals")

if not has_patterns:
    print("\n❌ RESIDUALS DON'T HAVE STRONG TEMPORAL PATTERNS")
    print("LSTM won't help much. XGBoost alone is likely optimal.")
    print("Consider these alternatives:")
    print("1. Feature engineering for XGBoost")
    print("2. Ensemble of different models")
    print("3. Hyperparameter tuning for XGBoost")
else:
    print("\n✅ RESIDUALS HAVE TEMPORAL PATTERNS - LSTM MIGHT HELP")

# 3. CORRECT LSTM approach if patterns exist
if has_patterns:
    print("\n=== TRAINING LSTM ON RESIDUALS (CORRECT APPROACH) ===")
    
    # Scale only training residuals
    scaler = StandardScaler()
    residual_train_scaled = scaler.fit_transform(residual_train.values.reshape(-1, 1))
    
    # Create sequences from training residuals only
    def create_sequences(data, window_size=5):  # Smaller window
        X, y = [], []
        for i in range(window_size, len(data)):
            X.append(data[i-window_size:i])
            y.append(data[i])
        return np.array(X), np.array(y)
    
    X_lstm_train, y_lstm_train = create_sequences(residual_train_scaled)
    
    # Split training data for validation
    split_idx = int(0.8 * len(X_lstm_train))
    X_train_val = X_lstm_train[:split_idx]
    y_train_val = y_lstm_train[:split_idx]
    X_val = X_lstm_train[split_idx:]
    y_val = y_lstm_train[split_idx:]
    
    # Convert to tensors
    X_train_tensor = torch.FloatTensor(X_train_val)
    y_train_tensor = torch.FloatTensor(y_train_val)
    X_val_tensor = torch.FloatTensor(X_val)
    y_val_tensor = torch.FloatTensor(y_val)
    
    # Simple LSTM model
    class SimpleLSTM(nn.Module):
        def __init__(self, input_size=1, hidden_size=32, num_layers=1):
            super(SimpleLSTM, self).__init__()
            self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
            self.fc = nn.Linear(hidden_size, 1)
            self.dropout = nn.Dropout(0.2)
        
        def forward(self, x):
            out, _ = self.lstm(x)
            out = self.dropout(out[:, -1, :])
            out = self.fc(out)
            return out
    
    model = SimpleLSTM()
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
    
    # Training with validation
    train_losses = []
    val_losses = []
    
    print("Training LSTM on residuals...")
    for epoch in range(100):
        # Training
        model.train()
        optimizer.zero_grad()
        output = model(X_train_tensor)
        train_loss = criterion(output.squeeze(), y_train_tensor.squeeze())
        train_loss.backward()
        optimizer.step()
        
        # Validation
        model.eval()
        with torch.no_grad():
            val_output = model(X_val_tensor)
            val_loss = criterion(val_output.squeeze(), y_val_tensor.squeeze())
        
        train_losses.append(train_loss.item())
        val_losses.append(val_loss.item())
        
        if epoch % 20 == 0:
            print(f'Epoch {epoch}, Train Loss: {train_loss.item():.6f}, Val Loss: {val_loss.item():.6f}')
    
    # Plot training curves
    plt.figure(figsize=(10, 4))
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.title('LSTM Training on Residuals')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.yscale('log')
    plt.show()
    
    # For prediction on test set, we need to create a rolling prediction
    # This is tricky because we need residuals to predict residuals
    print("\n=== MAKING PREDICTIONS ===")
    print("Note: For test set, we can't use future residuals.")
    print("We'll use a rolling prediction approach.")
    
    # Use last training window to start predictions
    model.eval()
    window_size = 5
    
    # Start with last training residuals
    current_window = residual_train_scaled[-window_size:].copy()
    lstm_predictions = []
    
    # Rolling prediction for test set
    for i in range(len(y_test)):
        with torch.no_grad():
            input_tensor = torch.FloatTensor(current_window.reshape(1, window_size, 1))
            pred = model(input_tensor).item()
            lstm_predictions.append(pred)
            
            # Update window with prediction (not ideal, but necessary)
            current_window = np.roll(current_window, -1)
            current_window[-1] = pred
    
    # Transform back to original scale
    lstm_predictions = np.array(lstm_predictions)
    lstm_predictions_original = scaler.inverse_transform(lstm_predictions.reshape(-1, 1)).flatten()
    
    # Final combined prediction
    final_predictions = xgb_pred_test + lstm_predictions_original
    
    # Evaluate
    xgb_rmse = np.sqrt(mean_squared_error(y_test, xgb_pred_test))
    combined_rmse = np.sqrt(mean_squared_error(y_test, final_predictions))
    
    print(f"\nResults:")
    print(f"XGBoost RMSE: {xgb_rmse:.4f}")
    print(f"Combined RMSE: {combined_rmse:.4f}")
    print(f"Improvement: {((xgb_rmse - combined_rmse) / xgb_rmse * 100):.2f}%")

else:
    print("\n=== ALTERNATIVE APPROACHES ===")
    print("Since residuals don't have temporal patterns, try:")
    
    # 1. XGBoost hyperparameter tuning
    print("\n1. XGBoost Hyperparameter Tuning:")
    from xgboost import XGBRegressor
    from sklearn.model_selection import RandomizedSearchCV
    
    param_dist = {
        'n_estimators': [100, 200, 300, 500],
        'max_depth': [3, 4, 5, 6, 7],
        'learning_rate': [0.01, 0.05, 0.1, 0.2],
        'subsample': [0.8, 0.9, 1.0],
        'colsample_bytree': [0.8, 0.9, 1.0]
    }
    
    xgb_tuned = XGBRegressor(random_state=42)
    random_search = RandomizedSearchCV(
        xgb_tuned, param_dist, n_iter=20, cv=5, 
        scoring='neg_mean_squared_error', random_state=42
    )
    
    print("Performing hyperparameter tuning...")
    random_search.fit(X_train, y_train)
    
    # Best model predictions
    best_xgb = random_search.best_estimator_
    tuned_pred = best_xgb.predict(X_test)
    tuned_rmse = np.sqrt(mean_squared_error(y_test, tuned_pred))
    
    print(f"Original XGBoost RMSE: {np.sqrt(mean_squared_error(y_test, xgb_pred_test)):.4f}")
    print(f"Tuned XGBoost RMSE: {tuned_rmse:.4f}")
    print(f"Best parameters: {random_search.best_params_}")
```
